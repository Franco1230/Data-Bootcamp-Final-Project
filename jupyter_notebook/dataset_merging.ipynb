```python
import pandas as pd
```

```python
house_df=pd.read_csv('Resources/house.csv')
school_df= pd.read_csv('Resources/school.csv')
crime_df= pd.read_csv('Resources/crime.csv')
```

```python
crime_per_postcode= {}

for p in crime_df['Postcode']:
    if p not in crime_per_postcode.keys():
        crime_per_postcode[p] = 1
    else: 
        crime_per_postcode[p] += 1
```

```python
house_df['CrimeRate'] = 0

for row in range(len(house_df)):
      
    if house_df.loc[row,'Postcode'] in crime_per_postcode.keys():
        house_df.loc[row,'CrimeRate'] = crime_per_postcode[house_df.loc[row,'Postcode']]
        
```

```python
school_per_postcode= {}

for p in school_df['Postcode']:
    if p not in school_per_postcode.keys():
        school_per_postcode[p] = 1
    else: 
        school_per_postcode[p] += 1
```

```python
house_df['NearbySchools'] = 0

for row in range(len(house_df)):
      
    if house_df.loc[row,'Postcode'] in school_per_postcode.keys():
        house_df.loc[row,'NearbySchools'] = school_per_postcode[house_df.loc[row,'Postcode']]
        
```

```python
house_df.to_csv('Resources/house_crime_school.csv')
```
