```python
# Dependencies and Setup
import pandas as pd
```

```python
# Set source file path
merged_data = "../source_data/house_crime_school.csv"

# Set output file path
cleaned_data = "../cleaned_data/cleaned_house_crime_school.csv"
```

```python
df = pd.read_csv(merged_data)
df
```

```python
cleaned_df = df.dropna(subset = ["Price"])
cleaned_df = cleaned_df.drop(["Unnamed: 0", "Method", "SellerG", ], axis = "columns")
cleaned_df["Date"] = pd.to_datetime(df["Date"], format = "%d/%m/%Y")
cleaned_df["Price"] = cleaned_df["Price"].str.replace(",", "").astype(float)
cleaned_df = cleaned_df[cleaned_df["Landsize"] !=0]
cleaned_df = cleaned_df.reset_index(drop = True)
```

```python
cleaned_df
```

```python
cleaned_df.to_csv(cleaned_data, index = False, encoding = "utf-8")
```
