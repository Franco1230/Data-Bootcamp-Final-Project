```python
# Dependencies and Setup
import pandas as pd
```

```python
# Set source file path
merged_data = "../cleaned_data/cleaned_house_crime_school.csv"
predict_data = "../output/price_prediction_df.csv"

# Set output file path
merged_predict_data = "../output/predicted_cleaned_house_crime_school.csv"
```

```python
df = pd.read_csv(merged_data)
df
```

```python
df = df.sort_values("Price")
df = df.reset_index(drop = True)
df = df.reset_index()
df
```

```python
predict_df = pd.read_csv(predict_data)
predict_df
```

```python
predict_df = predict_df.sort_values("Actual")
predict_df = predict_df.rename({"Actual": "Price"}, axis = "columns")
predict_df = predict_df.reset_index(drop = True)
predict_df = predict_df.reset_index()
predict_df = predict_df.round({"Prediction": 2})
predict_df
```

```python
merged_predict_df = pd.merge(df, predict_df, on = "index")
merged_predict_df = merged_predict_df.drop(["index", "Price_x"], axis = "columns")
merged_predict_df = merged_predict_df.rename({"Price_y": "Price"}, axis = "columns")
merged_predict_df
```

```python
merged_predict_df.to_csv("../output/merged_prediction.csv")
```
